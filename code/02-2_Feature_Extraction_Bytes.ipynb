# N-Gram

```python
def nGram(file_bytes, n):
    
    ngram_bytes = {}
    
    for i in range(len(file_bytes) - (n - 1)):
        
        seq = ".".join(([str(v) for v in file_bytes[i:i + n]]))
        
        try:
            ngram_bytes[seq] += 1
        except:
            ngram_bytes[seq] = 1
    
    return ngram_bytes
```

# N-Gram : whole line bytes

```python
import numpy as np
```

```python
def wholeLineBytes(file_bytes):
    
    ngram_bytes = {}
    
    for line in np.resize(file_bytes, (len(file_bytes) // 16, 16)):
        
        seq = '.'.join([str(v) for v in line])
        
        try:
            ngram_bytes[seq] += 1
        except:
            ngram_bytes[seq] = 1
        
    return ngram_bytes
```

```python
sorted(wholeLineBytes(file_bytes).items(), key=lambda t: t[1], reverse=True)
```

# IMAGE

```python
import numpy as np
import mahotas
```

```python
def images(file_bytes):
    img_array = np.resize(file_bytes, (len(file_bytes) // 16, 16))
    
    b = int((img_array.shape[0] * 16) ** (0.5))
    b = 2 ** (int(np.log(b) / np.log(2)) + 1)
    a = int(img_array.shape[0] * 16 / b)
    
    img_array = img_array[:int(a * b / 16), :]
    img_array = np.reshape(img_array, (a, b))
    
    colnames = ['imgHaralick_' + str(i) for i in range(52)]
    
    features = np.squeeze(mahotas.features.haralick(img_array))
    features = [d for v in features for d in v]
    
    feature_list = dict()
    
    for i in range(len(colnames)):
        feature_list[colnames[i]] = features[i]
    
    return feature_list
```
